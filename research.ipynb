```python
import pdfplumber
from pathlib import Path
```

```python
data_path = Path("data/Data Input.pdf")
```

```python
def extract_pdf(
        path: Path = data_path
):
    with pdfplumber.open(path) as file:
        return "".join([p.extract_text() for p in file.pages if p.extract_text()])
```
